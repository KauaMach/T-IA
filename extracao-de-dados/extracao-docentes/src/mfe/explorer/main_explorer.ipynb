```python
from collections import Counter
from pathlib import Path

FILES_TO_RENAME_DIR = Path("../../../data/input/raw/files_to_rename")
```

```python
extensions = Counter()

for file in FILES_TO_RENAME_DIR.rglob("*"):
    if file.is_file():
        ext = file.suffix.lower() if file.suffix else "(Without extension)"
        extensions[ext] += 1

print(f"Total de arquivos: {sum(extensions.values())}")
print(f"Extensões encontradas ({len(extensions)}):\n")
for ext, count in extensions.most_common():
    print(f"  {ext:<20} {count:>5} arquivo(s)")
```

```python
import pandas as pd

df_extensions = pd.DataFrame(
    extensions.most_common(), columns=["format", "quantity"]
).assign(percentual=lambda df: (df["quantity"] / df["quantity"].sum() * 100).round(2))

df_extensions
```

```python
import json

all_extensions = set()
for file in FILES_TO_RENAME_DIR.rglob("*"):
    if file.is_file():
        ext = file.suffix.lower() if file.suffix else "(Without extension)"
        all_extensions.add(ext)
print(f"Total de extensões únicas: {len(all_extensions)}")
with open("distinct_extensions.json", "w") as f:
    json.dump(sorted(all_extensions), f, indent=2)
```
